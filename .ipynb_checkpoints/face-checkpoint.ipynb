```python
import tensorflow as tf

# Charger le modèle CNN entraîné
model = tf.keras.models.load_model('C:/Users/bouaz/Downloads/emotions/modelv3.keras')
import numpy as np
```

```python
import cv2

# Charger le classificateur de visage Haar Cascade
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Démarrer la capture vidéo
cap = cv2.VideoCapture(0)

while True:
    # Lire une frame de la webcam
    ret, frame = cap.read()
    
    # Convertir en niveaux de gris pour la détection de visage
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    # Détecter les visages dans l'image
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)
    
    # Dessiner un rectangle autour du visage détecté
    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x+w, y+h), (255, 0, 0), 2)
        
        # Extraire la région du visage
        face_roi = gray[y:y+h, x:x+w]
        
        # Redimensionner l'image du visage pour qu'elle corresponde à l'entrée du modèle
        resized_face = cv2.resize(face_roi, (48, 48))  # Supposons que ton modèle attend des images de 48x48
        normalized_face = resized_face / 255.0
        reshaped_face = normalized_face.reshape(1, 48, 48, 1)  # Reshape pour correspondre à l'entrée du modèle
        
        # Faire une prédiction avec le modèle CNN
        emotion_prediction = model.predict(reshaped_face)
        emotion_label = np.argmax(emotion_prediction)
        
        # Afficher l'émotion détectée
        emotions = ['Colère', 'Dégoût', 'Peur', 'Joie', 'Tristesse', 'Surprise', 'Neutre']
        detected_emotion = emotions[emotion_label]
        
        cv2.putText(frame, detected_emotion, (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (36,255,12), 2)
    
    # Afficher la frame avec le visage détecté et l'émotion
    cv2.imshow('Détection d\'émotion', frame)
    
    # Quitter la boucle si la touche 'q' est pressée
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Libérer la capture et fermer les fenêtres
cap.release()
cv2.destroyAllWindows()
```
